```python
data = pd.read_csv('Data/agriculture_worldbank.csv')
```

```python
GDP = data.iloc[:, 1:]
contry = data.iloc[:, 0]
years  = data.columns[1:]
```

```python
plt.figure(figsize = (12, 7))
for gdp, country in zip(GDP.values, contry):
    plt.plot(years, gdp, label = country)

plt.xlabel('Years')
plt.ylabel('GDP')
plt.legend()
plt.show()
```
